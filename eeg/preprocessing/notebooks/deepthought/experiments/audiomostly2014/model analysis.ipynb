```python
import os;
from pylearn2.utils import serial;
model_file = os.path.join(MODEL_PATH, 'mlp_best.pkl');    
model = serial.load(model_file);  
```

```python
train_dataset = load_yaml(train_dataset_template)[0];
test_dataset = load_yaml(test_dataset_template)[0];
```

```python
import theano;
import numpy as np;
from pylearn2.space import CompositeSpace

def process_dataset(dataset) :
    layer = model.layers[0]
    data_specs = (CompositeSpace((
                                    layer.get_input_space(), 
                                    model.get_output_space())), # NOTE: Model - to get the labels
                               ("features", "targets"));
    minibatch = layer.get_input_space().make_theano_batch();
    output_fn = theano.function(inputs=[minibatch], outputs=layer.fprop(minibatch));
    it = dataset.iterator('sequential', batch_size=100, data_specs=data_specs);
    
    print layer.get_output_space()
    
    num_channels = layer.get_output_space().num_channels;
    print num_channels
    
    y_real = [];                
    inputs = [];
    output = [];
    for minibatch, target in it:
        out = output_fn(minibatch); # this hangs for convnet on Jeep2
        inputs.append(minibatch);
        output.append(out);
        y_real.append(np.argmax(target, axis = 1));
    #     break;
        
    inputs = np.vstack(inputs);    
    output = np.vstack(output);    
    y_real = np.hstack(y_real);   
    
    print inputs.shape
    print output.shape
    # print output[0, :, :, 0];
    return inputs, output, y_real;
```

```python
# too big -> > 3GB
import deepthought.util.fs_util as fs_util;
inputs, output, y_real = process_dataset(train_dataset);
fs_util.save(os.path.join(MODEL_PATH, 'cache', 'train_L0_output.pkl'), (inputs, output, y_real));
```

```python
def get_max_activations(inputs, output):
    max_activation = [[0., -1] for i in xrange(num_channels)];
    
    for i in xrange(len(inputs)):
        for c in xrange(num_channels):
            max_act = np.max(output[i, c, :, 0]);
            if max_act > max_activation[c][0]:
                max_activation[c] = [max_act, i];
    
    print max_activation;
        
    # print inputs[0].shape
    # print output[0].shape
    # print output[0][0, :, :, 0];
    
    return max_activation;
```

```python
sample_rate = 400
n_fft = 96
n_bins = 49
hop_length=24
import librosa
import matplotlib.pyplot as plt;
import matplotlib as mpl

import warnings
warnings.simplefilter('once', UserWarning)
warnings.simplefilter('default')
%matplotlib inline  

# avoid type3 fonts
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42

import deepthought.util.fs_util as fs_util;

def plot_frame(frame, spectrum_filepath=None, waveform_filepath=None):
    spectrum = frame.transpose();

    fig = plt.figure(figsize=(7, 5), dpi=600)
    axes = fig.add_subplot(111)
    im = librosa.display.specshow(spectrum, sr=sample_rate, hop_length=hop_length, x_axis='time', cmap='RdBu', vmin=-1, vmax=1)
    plt.colorbar(im)
    plt.ylabel('Frequency')
    fft_labels = librosa.core.fft_frequencies(sr=sample_rate, n_fft=n_fft)
    # plt.locator_params(axis='y',nbins=4)
    axes.set_yticks(xrange(0,n_bins))
    axes.set_yticklabels(fft_labels)
    for label in axes.yaxis.get_ticklabels():
        label.set_visible(False)
    for label in axes.yaxis.get_ticklabels()[::3]:
        label.set_visible(True)

    if spectrum_filepath is not None:
        fs_util.ensure_parent_dir_exists(spectrum_filepath);
        plt.savefig(spectrum_filepath, bbox_inches='tight')
    plt.show();

    waveform = librosa.core.istft(spectrum) #,hop_length=24, win_length=96, window=None)
    fig = plt.figure(figsize=(15, 3), dpi=600)
    axes = fig.add_subplot(111)
    plt.plot(waveform)
    plt.tight_layout()
    if waveform_filepath is not None:
        fs_util.ensure_parent_dir_exists(waveform_filepath);
        plt.savefig(waveform_filepath, bbox_inches='tight')
    plt.show();
```

```python
dataset_yaml = '''
    !obj:deepbeat.eeg.EEGDataset.EEGDataset {
        name : 'train',
        path : /Users/sstober/work/datasets/Dan/eeg/,
        suffix : _13goodchannels,
        subjects : [1],
        resample : [400,400],
        start_sample : 1600,
        stop_sample  : 11200,     # None (empty) = end of sequence 
        frame_size : 25,
        hop_size : 1,
        label_mode : rhythm_type,
    }
'''
dataset, dataset_yaml = load_yaml(dataset_yaml, {})
subject_sequences, subject_labels, channel_meta = load(dataset.datafiles[0]);

sample_rate = 400;
n_fft = 64 # 
hop_length = n_fft / 4;
spec = librosa.core.stft(subject_sequences[0], n_fft=n_fft, hop_length=hop_length)
spec = np.abs(spec)
spec = spec **2

fig = plt.figure(figsize=(200,6), dpi=600)
#ax = fig.add_subplot(111)
im = plt.imshow(spec, interpolation="none") #, cmap=blues_r);
#im = plt.matshow(spec, interpolation="none") #, cmap=blues_r);
#plt.colorbar(im)

fft_labels = librosa.core.fft_frequencies(sr=sample_rate, n_fft=n_fft)
plt.yticks(xrange(0,spec.shape[0]), fft_labels)

plt.savefig('spectrum.png', bbox_inches='tight')
```

```python
spec_cut = librosa.logamplitude(spec[0:12,:])
fig = plt.figure(figsize=(200,6), dpi=600)
im = librosa.display.specshow(spec_cut, sr=sample_rate, hop_length=hop_length, x_axis='time', cmap='afmhot')
plt.yticks(xrange(0,spec_cut.shape[0]), fft_labels)

plt.colorbar(im)

plt.savefig('spectrum3.png', bbox_inches='tight')
```

```python
# N consecutive overlapping frames of 256-point log power spectra 
# are *each* normalized to zero-mean and unit-variance to provide the input to the deep autoencoder.
# NOTE: N = input_length parameter in SDA or beat_pattern_width in CNN
from sklearn import preprocessing

n_freq_bins = 8;

cut_spec = spec[0:n_freq_bins, :]
log_spec = librosa.logamplitude(cut_spec)
scaled_spec = preprocessing.scale(log_spec)

#print scaled_spec.mean(axis=0)
#print scaled_spec.std(axis=0)

fig = plt.figure(figsize=(200,6), dpi=600)
im = librosa.display.specshow(scaled_spec, sr=sample_rate, hop_length=hop_length, x_axis='time', cmap='afmhot')
plt.yticks(xrange(0,scaled_spec.shape[0]), fft_labels)

plt.colorbar(im)

print scaled_spec[1]
plt.savefig('spectrum_scaled.png', bbox_inches='tight')
```

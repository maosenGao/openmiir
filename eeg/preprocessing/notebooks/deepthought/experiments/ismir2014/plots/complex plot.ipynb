```python
# TODO: plot a self-similarity matrix of frames

def annotate_group(name, xspan, ax=None):
    """Annotates a span of the x-axis"""
    def annotate(ax, name, left, right, y, pad):
        arrow = ax.annotate(name,
                xy=(left, y), xycoords='data',
                xytext=(right, y-pad), textcoords='data',
                annotation_clip=False, verticalalignment='top',
                horizontalalignment='center', linespacing=2.0,
                arrowprops=dict(arrowstyle='-', shrinkA=0, shrinkB=0,
                        connectionstyle='angle,angleB=90,angleA=0,rad=5')
                )
        return arrow
    if ax is None:
        ax = plt.gca()
    ymin = ax.get_ylim()[0]
    ypad = 0.01 * np.ptp(ax.get_ylim())
    xcenter = np.mean(xspan)
    left_arrow = annotate(ax, name, xspan[0], xcenter, ymin, ypad)
    right_arrow = annotate(ax, name, xspan[1], xcenter, ymin, ypad)
    return left_arrow, right_arrow

def make_second_bottom_spine(ax=None, label=None, offset=0, labeloffset=20):
    """Makes a second bottom spine"""
    if ax is None:
        ax = plt.gca()
    second_bottom = mpl.spines.Spine(ax, 'bottom', ax.spines['bottom']._path)
    second_bottom.set_position(('outward', offset))
    ax.spines['second_bottom'] = second_bottom

    if label is not None:
        # Make a new xlabel
        ax.annotate(label, 
                xy=(0.5, 0), xycoords='axes fraction', 
                xytext=(0, -labeloffset), textcoords='offset points', 
                verticalalignment='top', horizontalalignment='center')
```

```python
import math;
from prettyplotlib.colors import reds, blue_red, blues_r
from mpl_toolkits.axes_grid1 import make_axes_locatable

samples_per_second = 100; # = Hz
beat_length_ms = 160; # in ms
samples_per_beat = samples_per_second * beat_length_ms / 1000.0;
#print samples_per_beat
beats_per_bar = 12;
samples_per_bar = beats_per_bar * samples_per_beat;

## full version
#plot_max = 12801;
#x_max = 12798;
plot_max = 3201
x_max = 3198;

## small version
#plot_max = 2400;
#x_max = 2400;

x_max = min(x_max, len(x));

def add_heatmap(fig, heatmap_ax):
    heatmap_ax.set_xlim(xmin=0, xmax=plot_max) # this is essential, otherwise axis may misalign (different scaling)
    hm_len = plot_max;
    tmp = np.zeros([2,hm_len]) 
    tmp[:,0:x_max] = error[0:x_max]

    xr = np.linspace(0,hm_len-1,hm_len)
    yr = np.linspace(-1,1,2)
    #p = ax.pcolormesh(xr,yr, tmp, vmin=0, vmax=0.25, cmap=reds);
    p = heatmap_ax.pcolormesh(xr, yr, tmp, vmin=0, vmax=0.25, cmap=reds);

    divider = make_axes_locatable(plt.gca())
    cax = divider.append_axes("right", size=0.1, pad=-0.1) # HACK: placed INSIDE of plot to avoid breaking alignment of axes
    #cax = heatmap_ax.twinx();
    #cax.spines['right'].set_position(('outward', 18)) # NOTE: move first before changing something
    #cax.set_frame_on(True)
    #cax.patch.set_visible(False)
    fig.colorbar(p, cax=cax, label='abs. reconstruction error per sample', use_gridspec=True, orientation='vertical')
    
    #fig.colorbar(p, label='abs. reconstruction error per sample', orientation='vertical', pad=0.001)
    


def add_sample_axis(sample_axis):
    sample_axis.set_xlim(xmin=0, xmax=plot_max) # this is essential, otherwise axis may misalign (different scaling)
    # Drop the bottom spine by 40 pts
    sample_axis.spines['bottom'].set_position(('outward', 18)) # NOTE: move first before changing something
    sample_axis.xaxis.set_ticks(np.arange(0,plot_max+1,100));
    #sample_axis.set_xlabel(r"time / samples")
    # To make the border of the right-most axis visible, we need to turn the frame
    # on. This hides the other plots, however, so we need to turn its fill off.
    sample_axis.set_frame_on(True)
    sample_axis.patch.set_visible(False)

def add_time_axis(time_axis):
    time_axis.set_xlim(xmin=0, xmax=plot_max) # this is essential, otherwise axis may misalign (different scaling)
    ## set time axis labels
    time_ticks = [0];
    time_tick_labels = [0];    
    #time_axis.xaxis.set_ticks(np.arange(0,plot_max + 1,samples_per_second));    
    while time_ticks[-1]+samples_per_second < plot_max:
        time_ticks.append(time_ticks[-1]+samples_per_second);
        time_tick_labels.append(time_tick_labels[-1]+1);  
    for i in xrange(len(time_tick_labels)):
        time_tick_labels[i] = str(time_tick_labels[i])+'s';
    #print time_ticks;
    #print time_tick_labels
    time_axis.set_xticks(time_ticks)
    time_axis.set_xticklabels(time_tick_labels)
    #time_axis.set_xlabel(r"time (s)")
    time_axis.set_frame_on(True)
    time_axis.patch.set_visible(False)
    time_axis.xaxis.set_data_interval(0,plot_max)
    return time_axis;
    
def add_beat_axis(beat_axis):
    beat_axis.set_xlim(xmin=0, xmax=plot_max) # this is essential, otherwise axis may misalign (different scaling)
    ## set bar/beat axis labels
    beat_ticks = [0];
    beat_tick_labels = [0];
    while beat_ticks[-1]+samples_per_beat < plot_max:
        beat_ticks.append(beat_ticks[-1]+samples_per_beat);
        beat_tick_labels.append('');  
    for i in xrange(0, int(math.floor(len(beat_tick_labels) / beats_per_bar))):
        beat_tick_labels[i*beats_per_bar] = i+1;
    #print beat_ticks;
    #print beat_tick_labels;
    beat_axis.set_xticks(beat_ticks)
    beat_axis.set_xticklabels(beat_tick_labels)
    #beat_axis.set_xlabel(r"top: beats / bars")
    beat_axis.set_frame_on(True)
    beat_axis.patch.set_visible(False)
    return beat_axis;

# Annotate the groups
#groups = [('GroupA', (0, 100))]
#for name, xspan in groups:
#    annotate_group(name, xspan, ax=ax2)


#fig, ax = ppl.subplots(figsize=(200, 4), dpi=600)
#fig, axes = ppl.subplots(figsize=(20, 4), dpi=600) # small version


#fig, axes = ppl.subplots(2, figsize=(20, 8), dpi=600) # small version
#fig, axes = plt.subplots(nrows=2, figsize=(20,10)) # worked

#fig = plt.figure(figsize=(30,6), dpi=600) # small
fig = plt.figure(figsize=(200,6), dpi=600) # big

import matplotlib.gridspec as gridspec
gs = gridspec.GridSpec(2, 1,
                       width_ratios=[1,1],
                       height_ratios=[2,1],
                       hspace=0.3
                       )
#gs.update(hspace=0.00)

ax = plt.subplot(gs[0])

ax.plot(x[0:x_max], color='blue')
ax.plot(y[0:x_max], color='red')

plt.xlim(0,plot_max)
plt.ylim(-1,1)

add_heatmap(fig, ax);

sample_axis = ax;
time_axis = sample_axis.twiny(); # bottom
beat_axis = time_axis.twiny(); # top

add_sample_axis(sample_axis);
add_time_axis(time_axis); 
add_beat_axis(beat_axis); 

beat_axis.set_xlabel(r"top: beats / bars; bottom: time / samples")

#print sample_axis.get_xticks()
#print time_axis.get_xticks()
#print beat_axis.get_xticks()

## prepare date for activation plot

samples_per_frame = 72
hop_size = 6
frame_overlap = samples_per_frame / hop_size; # = 12
numframes = plot_max / hop_size;
#print numframes
padding = int(math.floor((samples_per_frame / 2.0) / hop_size));
#print padding

act_len = numframes - frame_overlap;
activations = np.zeros((15, numframes))
activations[:,padding:padding+act_len] = np.random.uniform(size=(15,act_len)) # FIXME: use real data

act_axis = plt.subplot(gs[1])
im = plt.imshow(activations, interpolation="none", extent=[0,numframes,0,1], aspect='auto')
#im = plt.imshow(activations, interpolation="none", extent=[0,numframes,0,1], aspect='auto', cmap=blues_r) # blue colormap
act_axis.set_xlabel(r"frames (size={}, hopsize={})".format(samples_per_frame, hop_size))
act_axis.set_yticks([]);

divider = make_axes_locatable(plt.gca())
cax = divider.append_axes("right", size=0.1, pad=-0.1) # HACK: placed INSIDE of plot to avoid breaking alignment of axes
fig.colorbar(im, cax=cax, label='activation', use_gridspec=True, orientation='vertical')

#plt.savefig('test.pdf')
plt.savefig('test.png', bbox_inches='tight')
```
